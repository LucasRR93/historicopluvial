```python
import pandas as pd
```

```python
df = pd.read_csv('inmet_filter.csv')
df.head(5)
```

```python
df['data'] = pd.to_datetime(df['datahora']).dt.date
df['ano'] = pd.to_datetime(df['datahora']).dt.year
df['mes'] = pd.to_datetime(df['datahora']).dt.month
df.head(5)
```

```python
df = df.drop(['datahora'], axis=1).groupby(['nomeEstacao', 'codEstacao', 'latitude', 'longitude', 'data', 'ano', 'mes']).sum().reset_index()
df.head(5)
```

```python
df['coordenadas'] = df.apply(lambda x : str(x['longitude']) + "," + str(x['latitude']), axis=1)
df.head(5)
```

```python
display(df.describe(include='all'))
display(df.dtypes)
```

```python
df['data'] = pd.to_datetime(df['data'])
```
